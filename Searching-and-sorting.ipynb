```
def selection_sort(l):
  for start in range(len(l)):
    minpos = start
    for i in range(start, len(l)):
      if l[i] < l[minpos]:
        minpos = i

    (l[start], l[minpos]) = (l[minpos], l[start])
```

```
def insertion_sort(l):
  for sliceend in range(len(l)):
    pos = sliceend
    while pos > 0 and l[pos] < l[pos - 1]:
      (l[pos], l[pos - 1]) = (l[pos - 1], l[pos])
      pos = pos - 1
```

```
def binary_search(l, high, low, v):
  (found, i) = (False, 0)
  if high - low == 1:
    (found, pos) = (False, -1)

  mid = (low + high) // 2
  if v == l[mid]:
    (found, pos) = (True, mid)
  elif v < l[mid]:
    return(binary_search(l, mid - 1, 0, v))
  elif v > l[mid]:
    return(binary_search(l, high, mid + 1, v))
  return(pos)
```

```
def merge(A, B):
  (C, m, n) = ([], len(A), len(B))
  (i, j) = (0, 0)
  while i + j < m + n:
    if i == m:
      C.append(B[j])
      j = j + 1
    elif j == n:
      C.append(A[i])
      i = i + 1
    elif A[i] <= B[j]:
      C.append(A[i])
      i = i + 1
    elif B[j] < A[i]:
      C.append(B[j])
      j = j + 1
  return(C)
```

```
def merge_sort(l, left, right):
  if right - left <= 1:
    return(l[left:right])
  elif right - left >= 1:
    mid = (right + left) // 2
    L = merge_sort(l, left, mid)
    R = merge_sort(l, mid, right)

  return(merge(L, R))
```

```
def quick_sort(l, left, right):
  if right - left <= 1:
    return()

  yellow = left + 1
  for green in range(left + 1, right):
    if l[green] <= l[left]:
      (l[yellow], l[green]) = (l[green], l[yellow])
      yellow = yellow + 1
  (l[left], l[yellow - 1]) = (l[yellow - 1], l[left])

  quick_sort(l, left, yellow - 1)
  quick_sort(l, yellow, right)
```
